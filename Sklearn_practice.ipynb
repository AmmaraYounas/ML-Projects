```python
from sklearn.model_selection import train_test_split
import pandas as pd
import numpy as np

np.random.seed(7)
```

#**Improving Model**

we can improve our model by using the hyperparameters
but the problem is we cant apply on trained data, its already trained and well set. We cant apply on test data because on which data we will be use for testing. So here's comes the **Validation Data** (separate splittion of data for validation or again training data for improving purposes).

So we split the data into three parts, **training 70%, validation 15% and testing 15%**

There are three way to improve:

## **1- Manually Improving**





```python
heart=pd.read_csv('/content/heart.csv')# single class classification problem
heart
```

```python
"""1- import data , defining x and y """

mixed_heart=heart.sample(frac=1)## shuffling it so that the patterns travels to different parts of data
x=mixed_heart.drop('target',axis=1)
y=mixed_heart['target']
```

```python
"""2- apply markers on data"""

train_split=round(0.7*len(mixed_heart))## to apply 70 % data to training
valid_split=round(train_split+0.15*len(mixed_heart))##15% data to valid after the training data

```

```python
"""3- splitting the data using markers"""

x_train,y_train=x[:train_split],y[:train_split]
x_valid,y_valid=x[train_split:valid_split],y[train_split:valid_split]
x_test,y_test=x[valid_split:],y[valid_split:]
```

```python
"""4- fitting the data and predict on validation data"""

from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score

gnb=GaussianNB()
gnb.fit(x_train,y_train)
predict_y=gnb.predict(x_valid)

print('accuracy_score = {}'.format(round(accuracy_score(y_valid,predict_y),2)))
print('precision_score= {}'.format(round(precision_score(y_valid,predict_y),2)))
print('recall_score= {}'.format(round(recall_score(y_valid,predict_y),2)))
print('f1_score= {}'.format(round(f1_score(y_valid,predict_y),2)))
```

```python
gnb.get_params()
```

```python
""" 5- applying the hyperparameters on validation data"""

gnb=GaussianNB(var_smoothing=0)
gnb.fit(x_train,y_train)
predict_y=gnb.predict(x_valid)

print('accuracy_score = {}'.format(round(accuracy_score(y_valid,predict_y),2)))
print('precision_score= {}'.format(round(precision_score(y_valid,predict_y),2)))
print('recall_score= {}'.format(round(recall_score(y_valid,predict_y),2)))
print('f1_score= {}'.format(round(f1_score(y_valid,predict_y),2)))


## the result has not improved although the hyperparameters are applied this way
```

## **2 Using RSCV (Randomized Search Cross Validation)**

```python
"""1-initializing model"""

from sklearn.ensemble import RandomForestClassifier

rscv_clf=RandomForestClassifier(n_jobs=1)## n_jobs for the processor to run the program single handed which runs programs parallely
rscv_clf.get_params()
```

```python
"""2- Splitting the data into train and test"""
x_train,x_test,y_train,y_test =train_test_split(x,y,
                                test_size=0.2)
param_dic={'max_depth':[None,5,10,20,30],
           'n_estimators':[10,100,200,500,1000,2000],
           'min_samples_split':[2,4,6],
           'min_samples_leaf':[1,2,4],
           'max_features':['auto','sqrt']}## defining dictionary for rscv parameter
```

```python
"""3- run the function rscv """
from sklearn.model_selection import RandomizedSearchCV

clf=RandomizedSearchCV(estimator= rscv_clf,
                       param_distributions=param_dic,
                       verbose=2,
                       cv=5,
                       n_iter=10)

"""4- training using this new model"""
clf.fit(x_train,y_train)
```

```python
## to know the value of hyperparameters """
clf.best_params_
```

```python
"""5- Predict the model"""

predict_y=clf.predict(x_test)

print('accuracy_score = {}'.format(round(accuracy_score(y_test,predict_y),2)))
print('precision_score= {}'.format(round(precision_score(y_test,predict_y),2)))
print('recall_score= {}'.format(round(recall_score(y_test,predict_y),2)))
print('f1_score= {}'.format(round(f1_score(y_test,predict_y),2)))

### this result is much much better than previous result
```
